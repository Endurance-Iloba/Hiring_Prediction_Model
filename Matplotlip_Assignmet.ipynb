```python
df = pd.read_csv('/content/drive/MyDrive/tips - tips (1).csv')
```

```python
#1. Plot()- Plots y and x as lines or markers

x = [1, 2, 3, 4]
y = [10, 15, 25, 30]

fig, ax = plt.subplots()
ax.plot(x, y, marker='o', color='maroon')
ax.set_title('Simple Line Plot')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
```

```python
#3. Scatter plot
fig, ax = plt.subplots()

sc = ax.scatter(df['total_bill'], df['tip'], c=df['total_bill'], cmap='gist_heat')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip Amount')
ax.set_xlim(0, 55)
ax.set_title('Total Bill vs Tip')

plt.colorbar(sc, ax=ax, label='Total Bill')
plt.show()
```

```python

df = pd.read_csv('/content/drive/MyDrive/tips - tips (1).csv')

#3. bar()- make a bar chart
days = df['day'].value_counts()

fig, ax = plt.subplots()
ax.bar(days.index, days.values, color='maroon')
ax.set_title('Number of Records per Day')
ax.set_xlabel('Day')
ax.set_ylabel('Count')
```

```python
#4. barh()

sizes = df['size'].value_counts()
fig, ax = plt.subplots()
ax.barh(sizes.index, sizes.values, color='maroon')
ax.set_title('Group Size Count')
ax.set_xlabel('Count')
ax.set_ylabel('Group Size')
```

```python
#5. hist()

fig, ax = plt.subplots()
ax.hist(df['total_bill'], bins=8, color='maroon')
ax.set_title('Total Bill')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
#6. boxplot()

fig, ax = plt.subplots()
ax.boxplot((df['tip']))
ax.set_title('Boxplot of Tips')
ax.set_ylabel('Tip Amount')
```

```python
#7. set_xlim() & set_ylim()

fig, ax = plt.subplots()
ax.scatter(df['total_bill'], df['tip'], color='maroon')
ax.set_xlim(0, 50)
ax.set_ylim(0, 12)
ax.set_title('Controlled Axis Limits')
```

```python
#8. Line plot

df_sorted = df.sort_values('total_bill')

colors = plt.cm.gist_heat(np.linspace(0.3, 0.9, 1))

fig, ax = plt.subplots()

ax.plot(
    df_sorted['total_bill'],
    df_sorted['tip'],
    color=colors[0]
)

ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip Amount')
ax.set_title('Size vs Total Bill')

plt.show()
```

```python
#9. The .add_axes() method brings in axes at an exact location
#[left, bottom, width, height] in figure coordinates (0 to 1)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1],)
x = np.linspace(0,10,100)
y = np.sin(x)
ax.plot(x,y, color='maroon')
```

```python
#10. Line chart

x=[10,20,30,40,50]
y=[50,100,150,200,250]
plt.plot(x,y, color='maroon')
plt.xlabel("Sales")
plt.ylabel("Quantity")
plt.title("Sales Report Line Chart")
```

```python
#11.twinx() adds a secondary axis

size=[2,3,4]
total_bill=[24.59,23.68,21.01]
fig,ax=plt.subplots()
ax.plot(size,total_bill, color='maroon', marker='*')
ax.set_title('size vs total bill')
ax.set_xlabel('size')
ax.set_ylabel('total bill')
plt.show()
```

```python
size=[2,3,4]
total_bill=[24.59,23.68,21.01]
tip=[3.61,3.31,3.50]
fig,ax1=plt.subplots()
ax1.plot(size,total_bill, color='maroon', marker='*')
ax1.set_xlabel('size')
ax1.set_ylabel('total bill')
ax2=ax1.twinx()
ax2.plot(size,tip, color='red', marker='+')
ax2.set_title('Does size affect tip?')
plt.show()
```

```python
#12.The .add_subplot () method allows for additional plot to be
#created in a grid layout

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig = plt.figure()
ax3 = fig.add_subplot(2, 2, 1)
ax3.plot(x, y, color='maroon')
ax3.set_title('Subplot 1')

ax4 = fig.add_subplot(2, 2, 2)
ax4.plot(x, np.cos(x), 'r')
ax4.set_title('Subplot 2')

ax5 = fig.add_subplot(2, 2, 3)
ax5.plot(x, np.tan(x), 'g')
ax5.set_title('Subplot 3')
ax5.set_ylim(-10, 10)

ax6 = fig.add_subplot(2, 2, 4)
ax6.plot(x, np.exp(-0.1*x)*np.sin(x), 'm')
ax6.set_title('Subplot 4')

plt.tight_layout()
plt.show()
```

```python
#13. ylim()-creates a limit to the y axis

fig,ax=plt.subplots()
ax.plot(x,y, color='maroon')
ax.set_ylim(0.2,0.5)
ax.set_title('ylim application')
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```

```python
#14. grid() it configures the axes grid lines
fig,ax=plt.subplots()
ax.grid(True)
ax.plot(x,y, color='maroon')
ax.set_ylim(0.2,0.5)
ax.set_title('ylim application')
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```

```python
#15. legend() it places a legend on the axes
fig,ax=plt.subplots()
ax.legend('Year')
ax.plot(x,y, color='maroon')
ax.set_ylim(0.2,0.5)
ax.set_title('ylim application')
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```

```python
#16.imshow() it displays an image on the axes
fig,ax=plt.subplots()
data=np.random.rand(5,5)
ax.imshow(data)
ax.plot(x,y, color='maroon')
ax.set_ylim(0.2,6)
ax.set_title('ylim application')
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```

```python
#17. pie() generates a piechart
sizes =[10,20,30,40]
my_colors=['red','pink','orange','maroon']
labels=['Fruits','Vegetables','Nuts','Meat']#slice names
explode=(0,0.2,0,0)#how much each slice pops out from center
fig,ax=plt.subplots()
ax.pie(sizes,labels=labels,explode=explode, colors=my_colors,shadow=True, startangle=90, autopct='%1.1f%%')
```

```python
#18. contour() plots contour lines
fig,ax=plt.subplots()
data=np.random.rand(5,5)
ax.contour(data)
ax.plot(x,y, color='maroon')
ax.set_ylim(0.2,6)
ax.set_title('ylim application')
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```

```python
#19. contourf() plot filled colours
fig,ax=plt.subplots()
data=np.random.rand(5,5)
ax.contourf(data, color='maroon')
ax.plot(x,y, color='maroon')
ax.set_ylim(0.2,6)
ax.set_title('ylim application')
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```

```python
#20.fill_between()fills the area between curves
fig,ax=plt.subplots()
ax.fill_between(x,y,hatch='///', color='maroon')
ax.plot(x,y)
ax.set_ylim(-1,8)
ax.set_title('ylim application')
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```
